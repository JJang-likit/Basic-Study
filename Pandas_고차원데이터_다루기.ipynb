# Pandas 정리 / 고차원 데이터 다루기

### DataFrame 새로운 칼럼 추가 및 삭제

```python
import pandas as pd

df4 = pd.DataFrame({'Class':['math', 'social', 'network', 'economy', 'big data'],
                 'Year':[2018, 2019, 2020, 2021, 2022],
                 'Price':[100, 150, 200, 250, 300],
                 'Location':['Korea', 'USA', 'China', 'Japan', 'India']},
                index = ['No1', 'No2', 'No3', 'No4', 'No5'])
df4
```

```python
# 새로운 칼럼 추가하기 - 내가 추가하고자 하는 칼럼명을 지정하고 데이터 값을 할당해준다.
# limitStudent 칼럼 추가하고 값을 30으로 지정

df4['limitStudent'] = 30
df4
```

```python
# numStudent 칼럼을 추가하고 값을 30, 40, 20, 10, 15로 지정
# 추가하고자 하는 칼럼명을 지정하고 데이터 값을 리스트로 할당해준다.

df4['numStudent'] = [30, 40, 20, 10, 15]
df4
```

```python
# priceLevel 칼럼을 추가하고 Price가 200과 같거나 크면 High, 작으면 Low를 할당해준다.
# High, Low값을 지정해주는 함수를 만들고 그 함수를 할당해준다.

def level(x):
  if x >=200:
    return 'High'
  else:
    return 'Low'
```

```python
# apply() -> 함수를 적용하기 위한 함수
df4['priceLevel'] = df4.apply(lambda x : level(x.Price), axis = 1)
df4
```

```python
# 기존 칼럼을 이용해서 새 칼럼 추가하기
# income 칼럼을 추가하고 Price와 numStudent의 값을 곱한 값으로 데이터를 할당해준다.

df4['income'] = df4.Price * df4.numStudent
df4
```

```python
# 기존 칼럼을 이용해서 새 칼럼 추가하기2
# 충원율계산 - 충원율은 정원(limitStudent) 대비 수강학생수(numStudent) 비율

df4['rate'] = df4.numStudent / df4.limitStudent *100
df4
```

```python
# 값을 반올림하고 싶은 경우 round() 함수를 사용해서 정리할 수 있다.

df4.round(0)

# 특정 칼럼만 지정하고 싶은 경우

# df4.round({'rate' : 0})
```

### 불필요한 칼럼 및 데이터 삭제하기

```python
# drop()을 사용해서 삭제가능
# 첫번째 인자는 삭제하고자 하는 인덱스명
# 두번째 인자는 axis를 나타냄
```

```python
# No5 삭제 (행 삭제)

df4_1 = df4.drop('No5', 0)
df4_1
```

```python
# PriceLevel 삭제(열 삭제)

df4_2 = df4.drop('priceLevel', 1)
df4_2
```

```python
# inplace를 True로 설정해주면 원본데이터에 반영됨 / 기본값은 False

df4.drop('priceLevel', 1, inplace = True)
df4
```

```python
# 여러 칼럼을 지우고 싶은 경우 인자를 리스트로 넣어주면 가능함

df4_3 = df4.drop(['Year', 'income'], 1)
df4_3
```

```python
# numStudent가 가장 작은 데이터를 찾아서 삭제하기

del_student = df4[df4.numStudent == df4.numStudent.min()]

df5 = df4.drop(del_student.index, 0)
df5
```

### 고차원 데이터 다루기

- 3차원 데이터를 기준으로 할 경우 학생별, 과목별, 연도별 3가지의 기준으로 데이터를 구성한다.
- 이런 식으로 고차원의 데이터를 구성이 가능하다.
- 고차원 데이터의 경우 칼럼 인덱스가 다중으로 되어 있다.

```python
from pandas import Series, DataFrame
import numpy as np
```

```python
# 샘플 데이터 만들기

np.random.seed(0)
df10 = pd.DataFrame(np.random.randint(50, 100, (5, 6)),
                    columns = [[2019, 2019, 2019, 2020, 2020, 2020],['English', 'Math', 'Science', 'English', 'Math', 'Science']],
                    index = ['Judy', 'Kelly', 'Bob', 'Teddy', 'Rick'])
df10.index.set_names('Names', inplace = True)
df10.columns.set_names(['Year', 'Subject'], inplace = True)
df10.loc['Rick', (2019, 'Science')] = np.nan
```

```python
df10
```

```python
df10.index
```

```python
df10.columns
```

```python
# 2020년 데이터만 선택하기

df10[2020]
```

```python
# 2020년의 영어 성적만 선택하기

df10[2020]['English']
```

```python
# 2020년의 영어 성적만 선택하기-2
# 튜플의 형태로 인자를 넣어줄 수 있다.

df10[(2020, 'English')]
```

```python
# Kelly의 성적만 선택
# 로우 인덱스를 색인하는 것이기 때문에 loc / iloc를 사용
# loc 의 경우 인덱스명, iloc 의 경우 인덱스의 숫자를 넣어주어야 한다.

df10.loc['Kelly']
```

```python
# Judy, Kelly, Bob의 성적만 선택하기
# 직접 리스트로 인자를 넣어줄 수도 있고 슬라이싱을 통해서도 색인 가능

df10.loc['Judy':'Bob']
```

```python
df10.iloc[:3]
```

```python
# Teddy와 Rick의 2019년 성적 선택하기 / df.loc[[로우인덱스], 칼럼인덱스]

df10.loc[['Teddy', 'Rick'], 2019]
```

```python
# Teddy와 Rick의 2020년 영어 성적 선택하기
# 리스트[ ]로 색인하게 되면 각각의 데이터를 색인하게 된다.
# 튜플( )의 경우 계층색인을 색인하게 된다.

df10.loc[['Teddy', 'Rick'], (2020, 'English')]
```

```python
# 2019년과 2020년의 영어 성적만 선택
# 1. 다중 칼럼을 색인하기 때문에 리스트로 인자를 넣어준다.
# 2. 계층색인을 색인해야하기 때문에 튜플 형태로 인자를 넣어준다.

df10[[(2019, 'English'), (2020, 'English')]]
```

```python
# 최상위 인덱스가 아닌 다른 인덱스를 찾는 방법
# xs(색인하고자 하는 인덱스 이름, axis, level)
# level은 어느 계층 인덱스를 찾는지 지정해줌 -> 최상위 인덱스는 0, 그 밑으로 1씩 증가

df10.xs('English', axis = 1, level = 1)
```

### 인덱스에 이름 부여하기 / set.names( )

```python
# 로우 인덱스의 이름을 'Student'로 정의하기

df10.index.set_names('Student', inplace = True)
df10
```

```python
# 각 컬럼의 이름 변경

df10.columns.set_names(['Years', 'Study'], inplace = True)
df10
```

### 인덱스 재구성
- 로우 인덱스를 칼럼 인덱스로 변경
- 로우 인덱스를 제거하고 칼럼으로 추가
- 로우 인덱스와 칼럼 인덱스를 서로 교환

```python
# 로우 인덱스를 칼럼으로 변경 / reset_index( )

df10.reset_index()
```

```python
# 인자로 받은 칼럼의 값들을 로우 인덱스로 변경

df10.reset_index().set_index('Student')
```

```python
 # swaplevel(index1, index2, axis) / 칼럼인덱스의 levle을 바꿔줌
 # index1과 index2가 로우 인덱스인 경우 axis = 0, 칼럼 인덱스인 경우 1

 df10.swaplevel('Years', 'Study', axis = 1)
```

```python
# sort.index(axis = 1)을 통해서 인덱스를 정렬할 수 있다.

df10.swaplevel('Years', 'Study', axis = 1).sort_index(axis =1)
```

```python
# stack() - 칼럼 인덱스를 로우 인덱스로 옮길 때 사용
# unstack() - 로우 인덱스를 칼럼 인덱스로 옮길 때 사용

df10.stack(1)
```

```python
df10.stack(0)
```

```python
df11 = df10.stack(1)
```

```python
# 연습 1 - Kelly의 성적만 선택하기

df11.loc['Kelly']
```

```python
# 연습 2 - Bob의 수학 성적만 선택
# 다중색인이기 때문에 튜플형태로 넣어줘야함

df11.loc[('Bob', 'Math')]
```

```python
# 연습 3 - 모든 학생들의 영어 성적만 선택
# 영어 성적은 로우 인덱스의 2번째 인덱스이기 때문에 xs함수를 사용해야함

df11.xs('English', axis = 0, level = 1)
```

```python
# 연습 4 - Teddy의 2020년 영어 성적만 선택

df11.loc[('Teddy', 'English'), 2020]
```

```python
# 연습 5 - 학생들의 과목별 성적의 평균을 구해서 새로운 칼럼 '평균'으로 저장

df11['avg'] = df11.mean(axis = 1)
df11
```

```python
# 연습 6 - 연도별 과목 평균을 구해서 출력
# 연도와 과목이 같은 칼럼에 있으면 계산하기 편하다.

df11 = df11.unstack(1)
df11
```
