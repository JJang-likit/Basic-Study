# Pandas 정리 / DataFrame

- DataFrame은 여러개의 시리즈의 묶음이다.
- 2차원 형태의 데이터(값, value)와 2개의 인덱스(index)로 구성되어 있다.

```python
import pandas as pd
from pandas import Series, DataFrame

print('finish')
```

```python
# 데이터프레임 생성하고 데이터 할당하기

df1 = DataFrame(data = [[1,2,3],
                  ['kity', 'bob', 'judy'],
                  [10.1, 10.2, 10.3]])
df1
```

```python
# 인덱스와 컬럼을 추가할 경우 / 인덱스 및 컬럼의 길이는 데이터의 모양과 개수와 일치해야 한다.

df1 = DataFrame(data = [[1,2,3],
                  ['kity', 'bob', 'judy'],
                  [10.1, 10.2, 10.3]],
                index = ['a', 'b', 'c'],
                columns = ['x', 'y', 'z'])
df1
```

```python
# 인덱스 및 컬럼 변경

df1.index = ['aa', 'bb', 'cc']
df1

df1.columns = ['xx', 'yy', 'zz']
df1
```

```python
# 딕셔너리 타입 데이터를 이용해서 데이터프레임 생성하기

people = {'Seoul': [100.5, 150.5, 200.5],
          'Tokyo': [50, 100, 150],
          'London': [120, 140, 160],
          'Paris': [85, 90, 95],
          'Katmandu': [30, 40, 55]}
```

### DataFrame 내가 원하는 데이터 선택하기

```python
df4 = pd.DataFrame({'Class':['math', 'social', 'network', 'economy', 'big data'],
                 'Year':[2018, 2019, 2020, 2021, 2022],
                 'Price':[100, 150, 200, 250, 300],
                 'Location':['Korea', 'USA', 'China', 'Japan', 'India']},
                index = ['No1', 'No2', 'No3', 'No4', 'No5'])
df4
```

```python
# 원하는 열(칼럼)만 조회하기
# Class만 조회

df4['Class']
```

```python
# 복수의 칼럼 조회하기 / 여러개의 칼럼을 찾는 경우 list로 담아서 넣어주어야 함
# Class와 Price 조회

df4[['Class', 'Price']]
```

#### 원하는 로우 인덱스 선택하기
- 데이터프레임에서 색인을 하게 되면 기본적으로 칼럼 인덱스를 선택하게 됨
- .loc / .iloc 를 사용해서 로우 인덱스를 색인할 수 있음
- loc와 iloc는 속성이기 때문에 함수에서 사용하는 axis를 사용할 수 없다.
- .loc 의 경우 로우 인덱스의 인덱스명을 사용해야 함 / DataFrame.loc['No1']
- .iloc 의 경우 로우 인덱스의 번호를 사용함 / DataFrame.iloc[0]

```python
# 원하는 행(로우)만 조회하기

df4.loc['No3']
```

```python
# iloc의 경우 숫자 인덱스를 사용하기 때문에 0부터 시작한다.

df4.iloc[2]
```

```python
df4.iloc[[2, 4]]
```

#### 인덱스 슬라이싱
- 범위 값을 주고 그 범위에 해당하는 데이터를 색인하는 방법
- 숫자, 문자 인덱스 모두 슬라이싱이 가능하다.
- 숫자 인덱스를 사용할 시 엔드 인덱스가 포함되지 않는다.
- 문자 인덱스를 사용할 시 엔드 인덱스가 포함된다.

```python
# 인덱스 슬라이싱
# No2~No4까지의 데이터 조회

df4.loc['No2' : 'No4']
```

```python
df4.iloc[1:3]
```

```python
# 여러개의 칼럼을 가지고 색인하고 싶은 경우

df4[['Year', 'Price']].loc[['No1', 'No2']]
# 먼저 색인하고 싶은 칼럼을 리스트로 넣어주고 같이 색인하고 싶은 칼럼을 loc를 통해서 추가로 넣어준다.
```

```python
df4.loc[['No3', 'No4'], ['Class', 'Location']]
```

```python
# 여러개의 칼럼을 슬라이싱 가능

df4.loc['No1':'No4', 'Class':'Price']
```

```python
# 칼럼인덱스만 가지고 슬라이싱을 하면 정확하게 데이터가 출력되지 않는다.

df4['Class':'Location']
```

```python
# loc는 기본적으로 로우 인덱스를 찾는다.
# 칼럼 인덱스를 사용해서 색인을 하려면 로우 인덱스도 같이 넣어주어야 한다.

df4.loc[: , 'Class':'Location'] # :은 모든 데이터를 조회하겠다는 의미
```

```python
# 조건 색인 / 특정 칼럼에 조건을 지정해주면 그 조건에 맞는 값을 불리언 값으로 보여준다.
# Price가 200 이하인 데이터 선택

df4['Price'] < 200
```

```python
# 리턴값을 프레임형식으로 받고 싶은 경우

df4[df4['Price'] < 200]
```
