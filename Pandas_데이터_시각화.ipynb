# Pandas_데이터 시각화

## 데이터 시각화
- Pandas의 Series와 DataFrame 객체는 plot( ) 함수를 제공해서 시각화를 할 수 있다.
- Pandas의 plot( ) 함수도 matplotlib을 기반으로 구현되어 있다.
- matplotlib에서는 plot(x), bar(x, height), hist(x,bins) 등의 함수가 Pandas에서는 plot( )함수로 시각화가 가능하다.

### DataFrame.plot( ) 함수 인자값
- X : x축으로 사용될 칼럼을 지정, 기본값은 로우 인덱스이다.
- Y : y축의 값으로 사용될 칼럼을 지정, 기본값은 수치형 칼럼이다.
- kind : 사용할 차트의 형태를 나타냄, 기본값은 line이다.
- subplots : 각각의 칼럼을 별도의 axes로 그릴지 여부를 나타냄, 기본값은 False로 지정되어 있다.
- figsize : figure의 크기, 튜플의 형태로 가로, 세로를 지정할 수 있다.
- title : axes의 제목, subplots = True인 경우 리스트의 형태로 지정할 수 있다.
- style : 칼럼별 라인 스타일을 지정할 수 있다.
- xticks, yticks : x축과 y축의 tick을 각각 지정할 수 있다.
- xlim, ylim : x축과 y축의 범위를 각각 지정할 수 있다.
- xlabel, ylabel : x축과 y축의 이름을 지정 할 수 있다.

```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시각화 연습

person = {'Korea' : [1053.5, 1023, 987],
          'Japan': [1023, 1067, 1123],
          'China': [512, 489, 487],
          'USA': [897, 872, 811],
          'India': [451, 421, 399]}

df = DataFrame(person)
df.index = [2018, 2019, 2020]
df
```

```python
# line 그래프 그리기 / matplotlib을 사용함

plt.plot(df)
```

```python
# line 그래프 그리기 2 / DataFrame.plot을 사용함

df.plot()
```

```python
# kind 인자를 사용해서 다양한 형태의 그래프를 그릴 수 있다.
# bar, barh, line 등 사용 가능

df.plot(kind = 'bar')
```

```python
# figsize 인자를 사용해서 크기를 변경할 수 있다.

df.plot(kind = 'line',
        figsize = (5, 3))
```

```python
 # title과 xticks 인자를 사용해서 제목과 x축의 인덱스명을 지정해줄 수 있다.

 df.plot(kind = 'line',
        figsize = (5, 3),
         title = 'person by years',
         xticks = [2018, 2019, 2020])
```

```python
# ylim 인자를 사용해서 y축의 범위를 지정할 수 있다.

df.plot(kind = 'line',
        figsize = (5, 3),
         title = 'person by years',
         xticks = [2018, 2019, 2020],
         ylim = [300, 1300])
```

```python
# xlabel, ylabel 인자를 통해서 x축과 y축에 제목을 지정할 수 있다.

df.plot(kind = 'line',
        figsize = (5, 3),
         title = 'person by years',
         xticks = [2018, 2019, 2020],
         ylim = [300, 1300],
        xlabel = 'year',
        ylabel = 'number') # xlable, ylabel은 pandas의 1.1 버전 이상에서만 가능하다.

        # pandas 버전이 낮은 경우 plt.xlable, plt.ylable 을 사용해서 x축, y축 제목을 지정할 수 있다.
```

```python
# style을 지정할 때는 칼럼의 개수 만큼 지정해주어야 한다.

df.plot(kind = 'line',
        figsize = (5, 3),
         title = 'person by years',
         xticks = [2018, 2019, 2020],
         ylim = [300, 1300],
        xlabel = 'year',
        ylabel = 'number',
        style = ['+-', 'x-', 'd-', 'o-', '<-'])
```

```python
# subplots을 True로 설정해주면 각각의 칼럼별로 그래프가 그려진다.

df.plot(kind = 'line',
        figsize = (5, 3),
         title = 'person by years',
         xticks = [2018, 2019, 2020],
         ylim = [300, 1300],
        xlabel = 'year',
        ylabel = 'number',
        style = ['+-', 'x-', 'd-', 'o-', '<-'],
        subplots = True)
```

```python
# DataFrame.T.plot( )을 사용하면 x축과 y축이 변경되여 시각화된다.
# 기존에는 연도가 x축이 되었지만 현재 그래프에서는 각 나라가 x축이 된다.

df.T.plot()
```

```python
df.T.plot(kind = 'bar', subplots = True, figsize = (5, 7))
```
